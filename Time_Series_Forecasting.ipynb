```python
import numpy as np
import pandas as pd
np.random.seed(42)
N = 100
rng = pd.date_range('2019-01-01', freq='D', periods=N)
df = pd.DataFrame(np.random.rand(N, 1), columns=['value'], index=rng)
df['value'][::7] = 10
```

```python
df
```

```python
from autots import AutoTS


# 创建 AutoTS 模型
model = AutoTS(forecast_length=10,
    frequency='D',
    prediction_interval=0.95,
    ensemble=None,
    max_generations=0,
    model_list = 'univariate',
    num_validations=1)

# 训练模型
model = model.fit(df['value'])
# 进行时间序列预测
prediction = model.predict(forecast_length=15)
```

```python
from autots.models.model_list import model_lists
print(model_lists.keys())
```

```python
# accuracy of all tried model results
model_results = model.results()
# and aggregated from cross validation
validation_results = model.results("validation")
validation_results
```

```python
# Print the details of the best model
print(model)
```
